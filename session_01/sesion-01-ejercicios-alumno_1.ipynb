## Nivel 1 — Llamada básica (obligatorio)

Envía un mensaje al modelo y muestra la respuesta por pantalla.

**Elige la sección de tu proveedor y ejecuta el código.**

### Opción A: OpenAI

```python
from openai import OpenAI
from google.colab import userdata

# Kimi usa la API de OpenAI pero con su propio endpoint
kimi_client = OpenAI(
    api_key=userdata.get("OPENAI_API_KEY"),  # tu key de Kimi está aquí
    base_url="https://api.moonshot.ai/v1"
)

response = kimi_client.chat.completions.create(
    model="kimi-k2.6",  # flagship de Kimi, comparable a GPT-5.4 / Claude Opus
    messages=[
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "user", "content": "Explain in two sentences what a REST API is."}
    ],
    temperature=1
)

print(response.choices[0].message.content)
```

### Opción B: Anthropic

```python
from openai import OpenAI
from google.colab import userdata

# DeepSeek usa la API de OpenAI con su propio endpoint
deepseek_client = OpenAI(
    api_key=userdata.get("ANTHROPIC_API_KEY"),  # tu key de DeepSeek está aquí
    base_url="https://api.deepseek.com/v1"
)

response = deepseek_client.chat.completions.create(
    model="deepseek-chat",  # V3 = strong general purpose, comparable a GPT-5.4
    messages=[
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "user", "content": "Explain in two sentences what a REST API is."}
    ],
    temperature=0.7
)

print(response.choices[0].message.content)
```

## Nivel 3 — Metadatos y coste (opcional)

Extrae de la respuesta del modelo:
- Tokens de entrada y salida
- Modelo utilizado
- Coste estimado de la llamada

Consulta las tarifas de tu proveedor:
- OpenAI: https://openai.com/api/pricing
- Anthropic: https://docs.anthropic.com/en/docs/about-claude/models

### Opción A: OpenAI

```python
response = kimi_client.chat.completions.create(
    model="kimi-k2.6",
    messages=[
        {"role": "system", "content": "You are a software project estimation expert with 20 years of experience. You respond in a direct and technical manner."},
        {"role": "user", "content": "Explain in two sentences what a REST API is."}
    ],
    temperature=1
)

print("=== RESPONSE ===")
print(response.choices[0].message.content)

print("\n=== METADATA ===")
print(f"Model: {response.model}")
print(f"Input tokens: {response.usage.prompt_tokens}")
print(f"Output tokens: {response.usage.completion_tokens}")
print(f"Total tokens: {response.usage.total_tokens}")

# kimi-k2.6 pricing (approximate, verify at platform.moonshot.cn)
INPUT_PRICE_PER_MILLION = 0.60
OUTPUT_PRICE_PER_MILLION = 2.50

input_cost = (response.usage.prompt_tokens / 1_000_000) * INPUT_PRICE_PER_MILLION
output_cost = (response.usage.completion_tokens / 1_000_000) * OUTPUT_PRICE_PER_MILLION
total_cost = input_cost + output_cost

print(f"\n=== ESTIMATED COST ===")
print(f"Input cost:  ${input_cost:.6f}")
print(f"Output cost: ${output_cost:.6f}")
print(f"Total cost:  ${total_cost:.6f}")
```
